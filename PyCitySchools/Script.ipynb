```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
school_data_to_load = Path("../Resources/schools_complete.csv")
student_data_to_load = Path("../Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of students
student_count = total_students = len(school_data_complete['Student ID'])
total_students
```

```python
# Calculate the total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
# average math score
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
# average reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
# percentage passing math
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# percentage passing reading
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# percentage passing overall
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a district summary DataFrame
district_summary = pd.DataFrame({
    "Total Schools": len(school_data),
    "Total Students": total_students,
    "Total Budget": total_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": passing_math_percentage,
    "% Passing Reading": passing_reading_percentage,
    "% Overall Passing Rate": overall_passing_rate},index=["District Summary"])

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map("{:.2f}%".format)

district_summary
```

```python
# School Summary
# School name and types
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
# list schools and total students per school
total_student_s = school_data.set_index('school_name')['size']
total_school_budget_s = school_data.set_index('school_name')['budget']
total_student_s
```

```python
# Total school Budget per school
total_school_budget_s = school_data.set_index('school_name')['budget']
total_school_budget_s
```

```python
# budget per capita per school
budget_per_student_s = total_school_budget_s/total_student_s
budget_per_student_s
```

```python
# average math scores per school
average_math_score_s = student_data.groupby('school_name')['math_score'].mean()
average_math_score_s
```

```python
# average reading scores per school
average_reading_score_s = student_data.groupby('school_name')['reading_score'].mean()
average_reading_score_s
```

```python
# percentage passing math per school
pass_math_s = student_data[student_data['math_score'] >= 70].groupby('school_name')['Student ID'].count()/total_student_s*100
pass_math_s
```

```python
# percentage passing reading per school
pass_read_s = student_data[student_data['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/total_student_s*100
pass_read_s
```

```python
# overall passing per school
overall_pass_s = student_data[(student_data['reading_score'] >= 70) & (student_data['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/total_student_s*100
overall_pass_s
```

```python
# school summary DataFrame

school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": total_student_s,
    "Per Student Budget": budget_per_student_s,
    "Total School Budget": total_school_budget_s,
    "Average Math Score": average_math_score_s,
    "Average Reading Score": average_reading_score_s,
    '% Passing Math': pass_math_s,
    '% Passing Reading': pass_read_s,
    "% Overall Passing": overall_pass_s
})
school_summary = school_summary[['School Type', 
                                 'Total Students', 
                                 'Total School Budget', 
                                 'Per Student Budget', 
                                 'Average Math Score', 
                                 'Average Reading Score',
                                 '% Passing Math',
                                 '% Passing Reading',
                                 '% Overall Passing']]

# formatting
school_summary.style.format({'Total Students': '{:,}',
                              "Total School Budget": "${:,.2f}",
                              "Per Student Budget": "${:.2f}",
                              'Average Math Score': "{:.2f}", 
                              'Average Reading Score': "{:.2f}", 
                              "% Passing Math": "{:.2f}", 
                              "% Passing Reading": "{:.2f}",
                              "% Overall Passing": "{:.2f}"})

school_summary
```

```python
# passing percentage ascending
school_summary.sort_values(["% Overall Passing"],ascending=True)
```

```python
# passing percentage decending
school_summary.sort_values(["% Overall Passing"],ascending=False, inplace=False)
```

```python
# grade level average math score for each school 
ninth_math = student_data[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
tenth_math = student_data[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleventh_math = student_data[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
twelfth_math = student_data[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

math_scores_by_grade = pd.DataFrame({
        "9th": ninth_math,
        "10th": tenth_math,
        "11th": eleventh_math,
        "12th": twelfth_math
})
math_scores = math_scores_by_grade[['9th', '10th', '11th', '12th']]
math_scores.index.name = "School Name"

# show and format
math_scores.style.format({'9th': '{:.2f}', 
                           "10th": '{:.2f}', 
                           "11th": "{:.2f}", 
                           "12th": "{:.2f}"})
```

```python
# grade level average reading scores for each school
ninth_reading = student_data[student_data['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
tenth_reading = student_data[student_data['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
eleventh_reading = student_data[student_data['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
twelfth_reading = student_data[student_data['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

reading_scores_by_grade = pd.DataFrame({
        "9th": ninth_reading,
        "10th": tenth_reading,
        "11th": eleventh_reading,
        "12th": twelfth_reading
})
reading_scores = reading_scores_by_grade[['9th', '10th', '11th', '12th']]
reading_scores.index.name = "School Name"

# show and format
reading_scores.style.format({'9th': '{:.2f}', 
                           "10th": '{:.2f}', 
                           "11th": "{:.2f}", 
                           "12th": "{:.2f}"})
```
